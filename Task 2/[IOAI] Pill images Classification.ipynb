```python
import os
import json
import shutil
```

```python
class_to_label_path = '/content/training/medicine2label.json'

with open(class_to_label_path, 'r') as f:
  class_to_label = json.load(f)

label_to_class = {v: k for k, v in class_to_label.items()}

print(label_to_class)
type(label_to_class)
```

```python
os.mkdir('/content/train')
os.mkdir('/content/val')

for label, name in label_to_class.items():
  if not os.path.exists(f'/content/train/{name}'):
    os.mkdir(f'/content/train/{name}')
  if not os.path.exists(f'/content/val/{name}'):
    os.mkdir(f'/content/val/{name}')
```

```python
label_file = '/content/training/labels.txt'

with open(label_file, 'r') as f:
  lines = f.readlines()

train_split = 0.9 * len(lines)
train_lines = lines[:int(train_split)]
val_lines = lines[int(train_split):]

train_label = {}
val_label = {}

for line in train_lines:
  name, label = line.strip().split(' ')
  train_label[name] = label_to_class[int(label)]

  shutil.move(f'/content/training/images/{name}', f'/content/train/{label_to_class[int(label)]}')

for line in val_lines:
  name, label = line.strip().split(' ')
  val_label[name] = label_to_class[int(label)]

  shutil.move(f'/content/training/images/{name}', f'/content/val/{label_to_class[int(label)]}')

print(train_label)
print(val_label)
```

```python
import yaml

data = {
    'train': '/content/train',
    'val': '/content/val',
    'nc': len(label_to_class),
    'names': list(label_to_class.values())
}

with open('data.yaml', 'w') as f:
  yaml.dump(data, f)
```

```python
import numpy as np
from ultralytics import YOLO

model = YOLO('/content/runs/classify/train/weights/best.pt')

test_images_path = '/content/public_test/images'

predictions = []

for image in os.listdir(test_images_path):
  image_path = os.path.join(test_images_path, image)
  result = model.predict(source = image_path)

  names_dict = result[0].names
  probs = result[0].probs.data.tolist()

  predicted_label = names_dict[np.argmax(probs)]
  predicted_index = class_to_label[predicted_label]

  predictions.append(f"{image} {predicted_index}")

  print(f'{image}: {predicted_label}')

with open('predictions.txt', 'w') as f:
    for prediction in predictions:
        f.write(prediction + '\n')
```
